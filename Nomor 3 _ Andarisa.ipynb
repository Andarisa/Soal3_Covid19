```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia'
x = requests.get(url)
datacorona = pd.read_html(x.text)

datacorona[3]
```

```python
df = datacorona[3].copy()
df = df.iloc[:35, :5]
df.columns = ['Provinsi', 'Confirmed', 'Recovered', 'Deaths', 'Active cases']
df.iloc[12]['Provinsi'] = 'Jakarta'

df[['Confirmed', 'Recovered', 'Deaths', 'Active cases']] = df[['Confirmed', 'Recovered', 'Deaths', 'Active cases']].astype('int64')
```

```python
df
```

```python
provinsi = pd.read_json("/Users/Andarisa/Desktop/Purwadhika/Modul 2/Provinsi.json")


provinsi
```

```python
data = pd.merge(df, provinsi, on='Provinsi')
data
```

```python
import folium as fo
```

```python
peta = fo.Map(
    location = [0.7893, 113.9213],
    zoom_start = 5
)

for i in range(len(data)):
    fo.Marker(
        [data.iloc[i]['latitude'], data.iloc[i]['longitude']],
        popup = '<b>Confirmed:' + str(data.iloc[i]['Confirmed']) + '</b><br><b>Recovered:' + str(data.iloc[i]['Recovered']) + '</b><br><b>Deaths:' + str(data.iloc[i]['Deaths']) + '</b></p>',
        tooltip = data.iloc[i]['Provinsi'],
    ).add_to(peta)


peta
```

```python
peta.save('Peta_Covid19.html')
peta
```
